# British Data Set

We begin with the British data set. Overall we have data for six turbines. In the following we focus on the second turbine because here benchmarks are given.

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Load Data

```python
# load data
turbine_2016 = pd.read_csv("../data/raw/British/Kelmarsh_SCADA_2016_3082/Turbine_Data_Kelmarsh_2_2016-01-03_-_2017-01-01_229.csv", header=9)
turbine_2017 = pd.read_csv("../data/raw/British/Kelmarsh_SCADA_2017_3083/Turbine_Data_Kelmarsh_2_2017-01-01_-_2018-01-01_229.csv", header=9)
turbine_2018 = pd.read_csv("../data/raw/British/Kelmarsh_SCADA_2018_3084/Turbine_Data_Kelmarsh_2_2018-01-01_-_2019-01-01_229.csv", header=9)
turbine_2019 = pd.read_csv("../data/raw/British/Kelmarsh_SCADA_2019_3085/Turbine_Data_Kelmarsh_2_2019-01-01_-_2020-01-01_229.csv", header=9)
turbine_2020 = pd.read_csv("../data/raw/British/Kelmarsh_SCADA_2020_3086/Turbine_Data_Kelmarsh_2_2020-01-01_-_2021-01-01_229.csv", header=9)
turbine_2021 = pd.read_csv("../data/raw/British/Kelmarsh_SCADA_2021_3087/Turbine_Data_Kelmarsh_2_2021-01-01_-_2021-07-01_229.csv", header=9)


# make sure that the data frames have the same columns
turbine_two = np.array([turbine_2016.columns, turbine_2017.columns, turbine_2018.columns, turbine_2019.columns, turbine_2020.columns, turbine_2021.columns])
assert np.all(turbine_two[:-1] == turbine_two[1:]), "The columns are not the same"

print(turbine_2016.keys())

# combine data and select first 27 columns
turbine_two = pd.concat([turbine_2016, turbine_2017, turbine_2018, turbine_2019, turbine_2020, turbine_2021])

# shape of data
print(turbine_two.shape)

# rename the '# Date and time' column to 'Date' for easier access, convert it to pd.datetime and set it as the index
turbine_two.rename(columns={'# Date and time': 'Date'}, inplace=True)
turbine_two['Date'] = pd.to_datetime(turbine_two['Date'])
turbine_two.set_index('Date', inplace=True)

turbine_two.head()
```

## 2. Data Visualization

```python
# summarize
turbine_two.describe()
```

```python
# show keys of data frame

for key in turbine_two.keys():
    print(key)
```

The keys that seem relevant are: Wind Speed, Wind direction, Nacelle position, Energy Export, Power, Rotor speed, Nacelle ambient temperature. 

```python
# extract relevant columns
turbine_two = turbine_two[['Wind speed (m/s)', 'Wind direction (Â°)', 'Nacelle position (Â°)', 'Energy Export (kWh)', 'Power (kW)',
                           'Nacelle ambient temperature (Â°C)', 'Rotor speed (RPM)']]

turbine_two.describe()
```

```python
# create plot with wind speed and energy export over time

start_date = '2017-01-01'
end_date = '2017-02-01'

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(turbine_two.loc[start_date:end_date, 'Wind speed (m/s)'], label='Wind speed (m/s)')
ax.set_ylabel('Wind speed (m/s)')
ax2 = ax.twinx()
ax2.plot(turbine_two.loc[start_date:end_date, 'Energy Export (kWh)'], label='Energy Export (kWh)', color='orange')
ax2.set_ylabel('Energy Export (kWh)')

# set title
ax.set_title('Wind speed and energy export over time')

# add legend
fig.legend(loc="upper left", bbox_to_anchor=(0,1), bbox_transform=ax.transAxes)

plt.show()
```

**Remark:** The behaviour of the energy exported is relatively well modeled by the wind speed. Together with the other predictors we should get a nice forecast, even though we do not have the data of the current timestep. This also can be seen in the correlation table

```python
# test correlation between the variables
turbine_two.corr()
```

The correlations displays the correlations one might expect.
